```python
from GenericScraper import GenericScraper
import GeneralPurposeTools as gpt
from bs4 import BeautifulSoup
import urllib3
import certifi
import requests
import time


from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
s_url = "https://sebgroup.com/pow/wave/apps/SwapRates/Default.aspx"#"https://sebgroup.com/large-corporates-and-institutions/prospectuses-and-downloads/rates/swap-rates"
```

```python
class SwapDataScraper(GenericScraper):
    def __init__(self, currency):
        self.currency = currency
        self.data = None
    
    def URLManufacture(self):
        return s_url 
    
    def Retreive(self):
        URL = self.URLManufacture()
        page = urllib3.PoolManager(cert_reqs='CERT_REQUIRED', ca_certs=certifi.where()).request('GET',URL)
        soup = BeautifulSoup(page.data, 'html.parser')
        self.data = soup
    
    def Scrape(self):
        currencySource = self.data.findAll('td', attrs={'style':'border-style: none'})#'class' :'normal1'
        allRows = self.data.findAll('tr', attrs={'class' :'normal1'})
        allHeaders = self.data.findAll('tr', attrs={'class' :'normal'})
        return currencySource, allRows, allHeaders
    
    def Display(self):
        pass
    
    def ProcessData(self):
        pass
    
    def Save(self):
        pass
```

```python
swapDataScraper = SwapDataScraper('SEK')
```

```python
swapDataScraper.Retreive()
curr, dataRows, headers = swapDataScraper.Scrape()
```

```python
for row in dataRows:
    valuesList = gpt.RemoveHTMLTags(str(row)).split('\r\n')
    for value in valuesList:
        valueInfo = value.split('  ')[-1]
        if not '\n' in valueInfo:
            print(valueInfo, len(valueInfo))
    print('-'*20)
```
